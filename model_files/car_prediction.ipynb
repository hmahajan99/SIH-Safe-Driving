```python
import numpy as np
import keras
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from keras import losses
from keras.models import Sequential
from keras.layers import Dense , Dropout
```

```python
model = Sequential()
```

```python
path_file = os.path.dirname(os.path.abspath('__file__')) + '/coordibnate_data.csv'
data_file = pd.read_csv(path_file)
```

```python
data_file.head()
```

```python
x = data_file.iloc[: , 0:9]
y = data_file.iloc[: , 9:]
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x , y)
```

```python
xtest.shape
```

```python
len(ytest) , len(ytrain)
```

```python
xtrain.shape
```

```python
model.add(Dense(128, kernel_initializer='normal',input_dim = xtrain.shape[1], activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dense(2, kernel_initializer='normal',activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mse'])
```

```python
model.fit(xtrain , ytrain , epochs = 200 , validation_data = (xtest , ytest))
```

```python
ypred = model.predict(xtest)
```

```python
ytest - ypred
```

```python
from keras.models import load_model

# Creates a HDF5 file 'my_model.h5'
model.save('my_model.h5')

del model
```
